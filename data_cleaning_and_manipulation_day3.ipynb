```python
east =df[df['Region'] == 'East']
south =df[df['Region'] == 'South']
west =df[df['Region'] == 'West']
north =df[df['Region'] == 'North']
```

```python
#when there are different column names, set axis = 1
#ignore index is so that the numbering starts from zero
alldf = pd.concat([east, south, west, north], axis=0, ignore_index=True)
alldf
```

```python
#reshuffling
alldf =alldf.sample(frac = 0.8, ignore_index=True)
alldf
```

```python
#THE APPLY METHOD:used for aggregate functions like mean, median and etc. it has effect on all numerical columns
alldf.apply ('mean', numeric_only = True) 
```

```python
alldf.groupby('Region').apply('max')
```

```python
#PIVOT TABLE


pv = pd.pivot_table(alldf,index='Region', columns='Type', values=['Sales', 'Units'])
pv
```

```python
#changing the default aggregate fxn
#on default, the aggregate fxn is mean
pv = pd.pivot_table(alldf,index='Region', columns='Type', values=['Sales', 'Units'], aggfunc='sum')
pv
```

```python
#margins work on any aggregate function you specify
pv = pd.pivot_table(alldf,index='Region', columns='Type', values=['Sales', 'Units'], aggfunc='sum', margins=True, margins_name='outcome')
pv
```

```python
#CROSS TABULATION: used to check the relationship btw 2 categorical columns **it is like a count
pd.crosstab(df['Region'], df['Type'], margins=True, margins_name='Total')
```
