```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Telecom Churn Rate Dataset.xlsx')
df.head()
```

```python
df['TotalCharges'].replace({' ':0}, inplace=True)
df.info()
```

```python
#histogram
plt.style.use('ggplot')
sns.displot(x=df['TotalCharges'], kde=True)
```

```python
#regression plot (regplot)
#regplot is only used to check the relationship btw 2 continuous (numeric) data
#relationship means that the values are greater then 0
sns.regplot(x=df['tenure'], y=df['TotalCharges'])
```

```python
sns.regplot(x=df['MonthlyCharges'], y=df['TotalCharges'])
```

```python
sns.scatterplot(x=df['tenure'], y=df['TotalCharges'])
```

```python
sns.jointplot(x=df['tenure'], y=df['TotalCharges'])
```

```python
sns.pairplot(df)
```

```python
#count plot is for categorical values 
sns.countplot(x=df['gender'])
```

```python
sns.countplot(x=df['gender'], hue=df['SeniorCitizen'])
```

```python
#bar plot is a comparism chart, where one of the axis must be categorical and the other must be numerical
sns.barplot(x=df['gender'], y=df['TotalCharges'])
```

```python
#the estimator on default is mean
sns.barplot(x=df['gender'], y=df['TotalCharges'], estimator=max)
```

```python
sns.barplot(x=df['gender'], y=df['TotalCharges'], estimator=max, hue=df['Churn'])
```

```python
#changing the colour of the bars. here, we don't use the command of color, we use violet
sns.barplot(x=df['gender'], y=df['TotalCharges'], estimator=max, palette='summer_r')
```

```python
sns.violinplot(x=df['gender'], y=df['TotalCharges'], estimator=max, hue=df['Churn'])
#the white dot in the middle is the median
```

```python
sns.violinplot(x=df['gender'], y=df['TotalCharges'], estimator=max, hue=df['Churn'], split=True)
```

```python
#swarmplot
sns.swarmplot(x=df['gender'], y=df['TotalCharges'],hue=df['Churn'])
```

```python
sns.stripplot(x=df['gender'], y=df['TotalCharges'],hue=df['Churn'])
```

```python
#boxplot: similar to histogram, however this one is used to identify outliers
#the black line is the median
sns.boxplot(y=df['MonthlyCharges'])
```

```python
sns.lineplot(x=df['MonthlyCharges'], y=df['tenure'])
```

```python
#heatmap
#1. you'll have to create a pivot table first
#2. heatmap is used to show graphical colour correlation
#fmt:to compress the values
df2= pd.pivot_table(df, columns='gender', index='Churn', values='TotalCharges')
df2
```

```python
sns.heatmap(df2, annot=True, fmt='0.7g')
```

```python
sns.clustermap(df2, annot=True, fmt='0.7g')
```
