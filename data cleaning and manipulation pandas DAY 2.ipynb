```python
import numpy as np
import pandas as pd
```

```python
#creating a dataframe
#creating a dataframe from list, tuple and numpy arrays
columns = ['firstca', 'secondca', 'thirdca']
ID = np.arange(1,51)
np.random.seed(2)
score = np.random.randint(5,10,150)
score = score.reshape(50,3)
df = pd.DataFrame(data = score, index = ID, columns =columns)
df
#df is a general variable name for dataframe
```

```python
#basic pandas operation
#selecting a column
df['firstca']
```

```python
#selecting multiple columns
df[['firstca', 'secondca']]
```

```python
#selecting a row using set index
df.loc[4]
```

```python
#selecting multiple rows using set index
df.loc[[4,8,15,17,25]]
```

```python
#selecting a row using default index; for default, you start counting from 0. 
#p.s the default is functional even if you set it
df.iloc[3]
```

```python
#selecting multiple rows using default index
df.iloc[[3, 7, 14, 16, 24]]
```

```python
#creating a new column using the insert method
age = np.random.randint(14,16,50).reshape(50,1)
df.insert(loc=0, column='Age', value=age)
df
```

```python
#creating a new column using the normal method; this method will add the new column to the end of the table
df['exams'] = np.random.randint(30,70,50).reshape(50,1)
df
```

```python
#creating a new column 
df['totalscore'] = df['firstca'] + df['secondca'] +df['thirdca'] + df['exams']
df
```

```python
#creating a row from existing row
df.loc['avg'] = (sum(df['Age']/len(df)),sum(df['firstca']/len(df)),sum(df['secondca']/len(df)),sum(df['thirdca']/len(df)),
                 sum(df['exams']/len(df)),sum(df['totalscore']/len(df)))
df
```

```python
#dropping existing rows and columns
```

df

```python
df.drop('totalscore', axis = 1, inplace =True) #axis is 1 for column
df.drop('avg', axis =0, inplace = True) #axis is 0 for row

df
```

```python
import requests
url_link = 'https://pokemondb.net/pokedex/all'
r = requests.get(url_link, headers ={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64;x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'})
read_html_pandas_data = pd.read_html(r.text)[0]
df_pokeman = pd.DataFrame(read_html_pandas_data)
```

```python
df_pokeman
```

```python
#saving to external file
df_pokeman.to_csv('pokeman.csv', index=False)
df_pokeman.to_excel('pokeman.xlsx', sheet_name ='GAME', index=False)
```

```python
with pd.ExcelWriter('POKEMANTYPE.xlsx', engine ='xlsxwriter') as writer:
    df_pokeman[df_pokeman['Type'] =='Water']. to_excel(writer, index=False, sheet_name='WATER', engine = 'xlsxwriter')
    df_pokeman[df_pokeman['Type'] =='Normal']. to_excel(writer, index=False, sheet_name='NORMAL', engine = 'xlsxwriter')
    df_pokeman[df_pokeman['Type'] =='psychic']. to_excel(writer, index=False, sheet_name='PSYCHIC', engine = 'xlsxwriter')
    df_pokeman[df_pokeman['Type'] =='Grass']. to_excel(writer, index=False, sheet_name='GRASS', engine = 'xlsxwriter')
    df_pokeman[df_pokeman['Type'] =='Fire']. to_excel(writer, index=False, sheet_name='FIRE', engine = 'xlsxwriter')
```
